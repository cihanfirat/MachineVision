```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt') 

print(f"YOLOv8- Nano Modeli Yüklendi. Eğitim Başlıyor...")

# 2. Eğitimi Başlat
results = model.train(
    data='data.yaml', 
    epochs=50,      
    imgsz=512,       
    batch=16,         # Bellek yönetimi için düşük tutuyoruz
    workers=4,
    device=0,        # GPU
    #patience=30,     # Erken durdurma
    #augment=True,    # Veri artırımı
    name='atik_yolov8_nano',
    exist_ok=True
)

print("\nEğitim Tamamlandı! Sonuçlar: runs/detect/atik_yolov8_nano/")
```

```python
from ultralytics import YOLO

# Eğitilmiş modeli yükle
model = YOLO("runs/detect/atik_yolov8_nano/weights/best.pt")

print("Model yüklendi. Tahmin yapılıyor...")

# Tek resimle tahmin
results = model.predict(
    source="test_pil.jpg",
    imgsz=512,
    conf=0.25,
    device=0,          # GPU
    save=True,         # Çıktıyı kaydet
    show=False         # Ekranda gösterme (istersen True yap)
)

print("Tahmin tamamlandı!")
print("Sonuçlar: runs/detect/predict/")
```
